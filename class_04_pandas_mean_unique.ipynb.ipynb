```python
import pandas as pd
df = pd.DataFrame ({
"Points" : [10,20,30,40,50],
"Team" : ["BARCELONA" , "MADRID" , "ATLETICO" , "SOCIEDAS" , "FC"]
 })
Total_points = df ["Points"].mean()
Team = df ["Team"].unique()
print (Total_points)
print (Team)
```
