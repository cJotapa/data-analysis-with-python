```python
import pandas as pd
df = pd.DataFrame({
    "productos": ["pan" , "arroz" , "leche" , "mecato" , "huevos"],
    "precio":[10,15,4,8,33],
    "cantidad_vendida":[100,400,300,40,500]
    })
#MOSTRAR TABLA COMPLETA
print ("DATA FRAME COMPLETO")
df ["ingresos_totales"]= df ["precio"] * df ["cantidad_vendida"]
print (df)

#ESTADISTICAS BASICAS
total_ingresos = df["ingresos_totales"].sum()
promedioeingresos= df ["cantidad_vendida"].mean()
max_cantidad = df["cantidad_vendida"].max()
min_cantidad = df["cantidad_vendida"].min()

#PRODUCTO CON EL MAYOR INGRESO
pmax = df.loc[df["cantidad_vendida"].idxmax()]

#PRODUCTO CON EL MENOR INGRESO
pmin = df.loc[df["cantidad_vendida"].idxmin()]
print ("--------------------------------------------------")
print ("RESUMEN")
print ("Total de ingresos:", total_ingresos)
print ("Promedio de ingresos:", promedioeingresos)
print ("Cantidad máxima vendida:", max_cantidad)
print ("Cantidad mínima vendida:", min_cantidad)
print ("--------------------------------------------------")
print ("Producto con mayor cantidad vendida:")
print (pmax)
print ("Producto con menor cantidad vendida:")
print (pmin)
```
